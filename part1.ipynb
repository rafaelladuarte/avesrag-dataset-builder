```python
import pandas as pd
```

```python
cbro = "data/treat_data/cbro.csv"
df_cbro = pd.read_csv(cbro,encoding="ISO-8859-1")
df_cbro.head()
```

```python
wikiaves = "data/treat_data/wikiaves.csv"
df_wikiaves = pd.read_csv(wikiaves)
df_wikiaves.head()
```

```python
df_wikiaves = df_wikiaves.dropna()
df_wikiaves
```

```python
df_wikiaves.iloc[1]
```

```python
df_cbro.loc[df_cbro["Nome do táxon"] == "Rhea americana"].iloc[0]
```

```python
df_cbro = df_cbro.merge(
    df_wikiaves[["font-blue-soft", "font-blue-soft href"]],
    left_on="Nome do táxon",
    right_on="font-blue-soft",
    how="left"
)
df_cbro.head()
```

```python
df_cbro = df_cbro.rename(columns={"font-blue-soft href": "url_wikiaves"})
df_cbro = df_cbro.drop(columns=["font-blue-soft"])
df_cbro.head()
```

```python
df_cbro = df_cbro.drop(columns=["id","Categoria"])
```

```python
df_cbro.head()
```

```python
df_cbro.rename(
    columns={
        'Nome do táxon': 'taxonomia',
        'Ordem':'ordem',
        'Família':'familia',
        'Subfamília': "subfamilia",
        'Gênero': 'genero',
        'Espécie': 'especie',	
        'Nome em Português': 'nome_pt',
        'English Name': "nome_eng"
    },
    inplace = True
)
df_cbro.head()
```

```python
df_cbro.info()
```

```python
df_cbro = df_cbro.dropna(subset=["url_wikiaves"])
```

```python
df_cbro.info()
```

```python
df_cbro.to_csv('data/treat_data/merge_cbro_wikiaves.csv')
```

```python
wikiaves_udi = "data/treat_data/wikiaves_uberlandia.csv"
df_wikiaves_udi = pd.read_csv(wikiaves_udi )
df_wikiaves_udi.head()
```

```python
df_wikiaves_udi.info()
```

```python
df_wikiaves_udi = df_wikiaves_udi[["font-blue-soft href","font-blue-soft"]]
df_wikiaves_udi.head()
```

```python
df_wikiaves_udi = df_wikiaves_udi.dropna()
df_wikiaves_udi.head()
```

```python
df_wikiaves_udi.info()
```

```python
df_cbro_udi = df_cbro[
    [
        'taxonomia','ordem','familia', "subfamilia",'genero',
        'especie',	 'nome_pt','nome_eng'
    ]
]

df_cbro_udi = df_cbro_udi.merge(
    df_wikiaves_udi[["font-blue-soft", "font-blue-soft href"]],
    left_on="taxonomia",
    right_on="font-blue-soft",
    how="left"
)
df_cbro_udi.head()
```

```python
df_cbro_udi.info()
```

```python
df_cbro_udi = df_cbro_udi.dropna(subset=["font-blue-soft href"])
df_cbro_udi.info()
```

```python
df_cbro_udi.rename(
    columns={
        'font-blue-soft href': 'url_wikiaves',
    },
    inplace = True
)
df_cbro_udi.head()
```

```python
df_cbro_udi = df_cbro_udi.drop(columns=["font-blue-soft"])
df_cbro_udi.head()
```

```python
df_cbro_udi.to_csv('data/treat_data/merge_cbro_wikiaves_uberlandia.csv')
```
